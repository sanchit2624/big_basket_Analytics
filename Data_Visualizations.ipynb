# DATA VISUALIZATION

Data visualization is a technique to represent the data in a form of Graphs, Plots and Charts.

There are majorly 3 libraries used for Data Visualization in python:
1. Matplotlib (Math Plotting)
2. Seaborn (Stats PLotting)
3. Plotly (Hover Library or Dashboarding)

# Matplotlib

- It is a Data Visualization library in Python
- Used for Mathematical Plotting
- With the help of Matplotlib we can also do resizing of a image in python.
- We basically use a sub-module or sub-library from matplotlib named as pyplot.

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('house pricing delhi- MagicBricks.csv')
```

```python
df
```

```python
# LINEPLOT...

# Line plot needs 2 axis (x and y) both numeric that will be more preferred.

x = np.array([1,2,3,4,5])
y = np.array([10,20,30,40,50])
```

```python
x,y
```

```python
# Create a line plot with these

plt.plot(x,y)
```

```python
# We can use markers to reflect the dat points or elements

plt.plot(x,y,marker='o')
```

```python
# Plotting data on line plot

plt.plot(df.Area,df.Price)
```

```python
plt.plot(df.Area,df.Price,marker='o')
```

```python
plt.plot(x,y,marker='o',linestyle='dashed',linewidth=2.5,color='k',markersize=10)
```

```python
# Data Visualization with Data Filteration...

a = df[df.Location=='Karol Bagh']
```

```python
a
```

```python
plt.plot(df[df.Location=='Karol Bagh']['Area'],df[df.Location=='Karol Bagh']['Price'],marker='o',ms=15)
```

```python
# It will resize the figure dimensional

plt.figure(figsize=(10,3))

plt.plot(df[df.Location=='Karol Bagh']['Area'],df[df.Location=='Karol Bagh']['Price'],marker='o',ms=15)
```

```python
# In Matplotlib there is no axis given we've to provide the labels accordingly

plt.figure(figsize=(10,3))
plt.plot(df.Area,df.Price,marker='o')
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.title('DELHI HOMES AREAS AND PRICE')
```

```python
# Scatter Plot

plt.scatter(df.Area,df.Price)
```

```python
# We can create comparison pots with scatter

maruti_speed = np.array([100,120,140,160,180])
maruti_avg = np.array([18,16,15,12,10])

ferrari_speed = np.array([250,270,300,330,350])
ferrari_avg = np.array([12,7,6,5,3])
```

```python
plt.figure(figsize=(10,4))
plt.scatter(maruti_speed,maruti_avg)
plt.scatter(ferrari_speed,ferrari_avg)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.scatter(maruti_speed,maruti_avg)
plt.scatter(ferrari_speed,ferrari_avg)
plt.grid()
plt.show()
```

```python
# Barplot

# Barplot is generated on a categorical data.

df.District.value_counts()
```

```python
plt.figure(figsize=(16,4))
col = ['g','orange','y','r','b','k','hotpink','purple','grey']
plt.bar(df.District.value_counts().index,df.District.value_counts(),color=col)
plt.xlabel('Delhi Disticts')
plt.ylabel('Total Available Homes')
```

```python
plt.figure(figsize=(16,4))
col = ['g','orange','y','r','b','k','hotpink','purple','grey']
plt.barh(df.District.value_counts().index,df.District.value_counts(),color=col)
plt.xlabel('Delhi Disticts')
plt.ylabel('Total Available Homes')
```
